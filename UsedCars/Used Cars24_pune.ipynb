# Used Car Web Scrapping 

## Importing the tools needed

```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

## Usedcars-Chennai

```python
url10='https://www.cars24.com/buy-used-car?sort=P&storeCityId=2423&pinId=411001'
driver.get(url10)


time.sleep(3)
```

```python
#Will keep scrolling down the webpage until it cannot scroll no more
last_height = driver.execute_script('return document.body.scrollHeight')
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(2)
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == last_height:
        break
    last_height = new_height
    #scrapping urls 
    car_url4=[]
    for i in driver.find_elements_by_xpath("//a[@class='_9Ue0B']"):
        car_url4.append(i.get_attribute("href"))
print(len(car_url4)) #printing length of urls
```

```python
#creating lists 

car_price10=[]
car_location10=[]
car_brand10=[]
car_fuel10=[]
car_drivKM10=[]
car_year10=[]
car_owners10=[]
car_year10=[]
car_hist10=[]
car_lastser10=[]
car_trans10=[]
car_reg10=[]
car_ins10=[]
```

```python
time.sleep(2)
for i in car_url4:
    driver.get(i)
    

    #Brand
    try:
        brand10=driver.find_element_by_xpath("//h2[@class='_2geSF']")
        car_brand10.append(brand10.text)
    except NoSuchElementException:
        car_brand10.append('N/A')
        
    #Car Driver Kilometer 
    try:
        driven_km10=driver.find_element_by_xpath("//li[@class='tHlIu'][3]/strong")
        car_drivKM10.append(driven_km10.text)
    except NoSuchElementException:
        car_drivKM10.append('N/A')

     #Fuel type 
    try:
        fuel10=driver.find_element_by_xpath("//li[@class='tHlIu'][4]/strong")
        car_fuel10.append(fuel10.text)
    except NoSuchElementException:
        car_fuel10.append('N/A')
        
    #Number of owners 
    try:
        own10=driver.find_element_by_xpath("//li[@class='tHlIu'][2]/strong")
        car_owners10.append(own10.text)
    except NoSuchElementException:
        car_owners10.append('N/A')
        
    #Year of Purchase
    try:
        year10=driver.find_element_by_xpath("//li[@class='tHlIu'][9]/strong")
        car_year10.append(year10.text)
    except NoSuchElementException:
        car_year10.append('N/A')
    
    #Car History
    try:
        hist10=driver.find_element_by_xpath("//li[@class='tHlIu'][1]/strong")
        car_hist10.append(hist10.text)
    except NoSuchElementException:
        car_hist10.append('N/A')
        
    #Last Service 
    try:
        serv10=driver.find_element_by_xpath("//li[@class='tHlIu'][6]/strong")
        car_lastser10.append(serv10.text)
    except NoSuchElementException:
        car_lastser10.append('N/A')
        
    #Registeration 
    try: 
        reg10=driver.find_element_by_xpath("//li[@class='tHlIu'][7]/strong")
        car_reg10.append(reg10.text)
    except NoSuchElementException:
        car_reg10.append('N/A')
        
    #Insurance
    try:
        ins10=driver.find_element_by_xpath("//li[@class='tHlIu'][8]/strong")
        car_ins10.append(ins10.text)
    except NoSuchElementException: 
        car_ins10.append('N/A')
        
    #Price
    try:
        price10=driver.find_element_by_xpath("//strong[@class='_2yYvS']")
        car_price10.append(price10.text)
    except NoSuchElementException:
        car_price10.append('N/A')
        
    #location
    try:
        loc10=driver.find_element_by_xpath("/html/body/div[1]/div[3]/div[4]/div/div/div/div[3]/div/div/strong")
        car_location10.append(loc10.text)
    except NoSuchElementException:
        car_location10.append("N/A")
```

```python
UsedCars_24=pd.DataFrame({})
UsedCars_24['Brand']=car_brand10
UsedCars_24['Driven Kilometers ']=car_drivKM10
UsedCars_24['Fuel']=car_fuel10
UsedCars_24['Location']=car_location10
UsedCars_24['Year of Purchase']=car_year10
UsedCars_24['Price']=car_price10
UsedCars_24['Car History']=car_hist10
UsedCars_24['Last Service']=car_lastser10
UsedCars_24['Registeration']=car_reg10
UsedCars_24['Insurance']=car_ins10
UsedCars_24
```

```python
UsedCars_24.to_csv("Cars24_pune.csv")
```
