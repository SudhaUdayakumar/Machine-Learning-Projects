# Pre-Processing Data

## Handling Missing Values

### Handling Missing Values for Object Columns with Mode

```python
df.info(verbose=True, memory_usage='deep', null_counts=True)
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

<span style='background:yellow'>**Observation:** We again listed the missing values and column types. So for the Classification columns, we can directly replace them with mode. Since the missing data% is less, we can replace them with mode, else we would have to consider if we need to retain or drop the column. Below we will replace with mode for missing classification data  </span>

```python

# Impute missing values for Gender
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)

# Impute missing values for Married
df['Married'].fillna(df['Married'].mode()[0],inplace=True)

# Impute missing values for Dependents
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)

# Impute missing values for SelfEmployed
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)

# Impute missing values for Credit_History
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

<span style='background:yellow'>**Observation:** We see that all classificaion columns, we have replaced the missing    </span>

### Handling Missing Values for Numerical Columns

<span style='background:pink'>**Note:** For Categorical columns, we were able to replace directly with mode. However for the numerical values, we need to check if they are normally distributed and the presence of outliers. If the columns are normally distributed and no outliers then we can replace with mean, else median  </span>

### LoanAmount Column

```python
plt.subplot(121)
sns.distplot(df['LoanAmount']);

plt.subplot(122)
df['LoanAmount'].plot.box(figsize=(16,5))

plt.show()
```

<span style='background:yellow'>**Observation:** Loan Amount is not normally distributed, we see a long tail on the right and the presence of outliers. So lets replace the missing values with median for LoanAmount  </span>

```python
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

### Loan_Amount_Term

```python
plt.subplot(121)
sns.distplot(df['Loan_Amount_Term']);

plt.subplot(122)
df['Loan_Amount_Term'].plot.box(figsize=(16,5))

plt.show()
```

<span style='background:yellow'>**Observation:** A clear picture of not a normally distributed curve and outliers, so lets replace with median  </span>

```python
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median(),inplace=True)
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

<span style='background:yellow'>**Observation:** We wee there are no more outliers in Loan_Amount_Term.  </span>

### Credit_History

```python
plt.subplot(121)
sns.distplot(df["Credit_History"]);

plt.subplot(122)
df['Credit_History'].plot.box(figsize=(16,5))

plt.show()
```

<span style='background:yellow'>**Observation:** Another column to be replaced with median and not mean, as not normally distributed and outliers presence  </span>

```python
df['Credit_History'].fillna(df['Credit_History'].median(),inplace=True)
```

```python
round(df.isnull().sum()/len(df)*100,2)
```

<span style='background:yellow'>**Observation:** All missing values have been handled.    </span>

# Importing Libraries for Model Building (Classification)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

## Train & Test Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=43)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
random_over_sampler = RandomOverSampler()
```

```python
x_res,y_res=random_over_sampler.fit_resample(x,y)
```

## Best Random State

```python
maxAccu=0
BestRs=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)    
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    f1=f1_score(y_test,pred,average="macro")
    if f1>maxAccu:
        maxAccu=f1
        BestRs=i
print("Best accuracy is ", maxAccu,"on Random_state  ",BestRs)
```

<span style='background:yellow'>**Observation:** We got our best random state, lets use it in our Model Building  </span>

## Instantiating the Models

```python
lg=LogisticRegression()
gnb=GaussianNB()
sv=SVC()
rf=RandomForestClassifier(n_estimators=100)
ad=AdaBoostClassifier(n_estimators=100)
dtc=DecisionTreeClassifier()
```

## Model Building and checking Evaluation Scores

```python
model=[lg,gnb,rf,ad,dtc,sv]

for m in model:
    m.fit(x_res,y_res)
    pred=m.predict(x_res)
    print("\n f1 score: of ",m)
    print(f1_score(y_res,pred))
    print("\n  Classificatin report:  of ",m)
    print(classification_report(y_res,pred))
    print("***************************************")
    score=cross_val_score(m,x_res,y_res,scoring="f1",cv=5)
    print(score)
    print("mean of cross val with scoring= accuarcy:",score.mean())
    print("***************************************")
    
```

<span style='background:yellow'>**Observation:** we have done 6 model building and we see in the classification report the F1-score is almost same for classes for 0 & 1.Random Forest & Decision tree, tops our chart here, lets compare with our cross validation in our next step    </span>

# Model Building and checking the Evaluation Scores

```python
modelList=[("Log Reg",lg),("Ada boost",ad),("Dec Tree Classifier", dtc),("Random Forest Class",rf),("Naive Bayes-G", gnb),("Support Vector",sv)]

def GetAccuracy(x,y,modelList):
    x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=.20,random_state=BestRs)
    ModelName= []
    modelf1=[]
    modelCrossval=[]
    f1CVDiff=[]
    for Nameofmodel, modelInstance in modelList:
        ModelName.append(Nameofmodel)
        modelInstance.fit(x_train,y_train)
        pred=modelInstance.predict(x_test)
        f1score= f1_score(y_test,pred,average="macro")
        modelf1.append(f1score)
        crossval=cross_val_score(modelInstance,x_res,y_res,scoring="f1",cv=5).mean()
        modelCrossval.append(crossval)
        f1CVDiff.append(f1score-crossval)
    df_acc=pd.DataFrame({"Name":ModelName,
                         "f1": modelf1,
                         "CrossValidation":modelCrossval,
                          "Difference in Accuracy": f1CVDiff})
    return df_acc
GetAccuracy(x_res,y_res,modelList)
```

<span style='background:yellow'>**Observation:** We have checked 6 models, and earlier we saw the F1-score in classification report was good for both classes 0 & 1. Here we calculate the F1 score and Cross validation, and then we find the diff. The. least diff can be used or the models with highest scores and least diff is used.
