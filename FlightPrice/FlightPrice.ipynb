# Problem Statement (Given)

Flight Price Prediction
Problem Statement:
Flight ticket prices can be something hard to guess, today we might see a price, check out the price of the same flight tomorrow, it will be a different story. We might have often heard travellers saying that flight ticket prices are so unpredictable. Here you will be provided with prices of flight tickets for various airlines between the months of March and June of 2019 and between various cities.

Size of training set: 10683 records

Size of test set: 2671 records

FEATURES:
Airline: The name of the airline.

Date_of_Journey: The date of the journey

Source: The source from which the service begins.

Destination: The destination where the service ends.

Route: The route taken by the flight to reach the destination.

Dep_Time: The time when the journey starts from the source.

Arrival_Time: Time of arrival at the destination.

Duration: Total duration of the flight.

Total_Stops: Total stops between the source and destination.

Additional_Info: Additional information about the flight

Price: The price of the ticket

 

You have to use your skills as a data scientist and build a machine learning model to predict the price of the flight ticket.

## Total_Stops column, truncating and assigning numerical values

```python
df["Total_Stops"].value_counts()
```

<span style='background:yellow'>**Observation:** non stop is actually, means "0" stops and eatch can be converted to its respective numerical values of 1/2/3/4   </span>

```python
def convert_into_stops(S):
    if S == '4 stops':
        return 4
    elif S == '3 stops':
        return 3
    elif S == '2 stops':
        return 2
    elif S == '1 stop':
        return 1
    elif S == 'non-stop':
        return 0
```

```python
df['Total_Stops'] = df['Total_Stops'].map(convert_into_stops)
```

```python
df["Total_Stops"].value_counts()
```

<span style='background:yellow'>**Observation:** Now we have converted them into numerical values and the value counts for the stops match before and after conversion. We have used this userdefined function, instead of Encoding, as we want to assign the correct value for 0, 1, 2,3 4, stops   </span>

## Converting Duration from Object column to Minutes column

```python
df.head()
```

<span style='background:yellow'>**Observation:** Duration e.g. is marked as 2h50m, the human mind will read it as 2 hours and 50 minutes. We will be able to calculate it to 170 minutes. However, it is an object column here. So we need to split the 2h 50m, into "2h" and "50m". And then convert the "2h" to 120 minutes. Then total the value. Conver the entire column to numerical format    </span>

```python

#Finding all numbers in a string
import re
duration_series = df["Duration"].str.findall(r'\d+')
```

```python
duration_series[:30]
```

<span style='background:yellow'>**Observation:** we see the hour and minutes are seperated.  </span>

```python
df.reset_index(inplace=True)
```

```python
from datetime import datetime
from datetime import timedelta
for i, duration_list in enumerate(duration_series.values):
    hours = int(duration_list[0])
    if len(duration_list)==1:
        mins = 0
    else:
        mins = int(duration_list[1])
        
    df.loc[i, "Duration"] = 60*hours + mins
    
```

```python
df.head()
```

<span style='background:yellow'>**Observation:** Now we see the duration of "2h 50m" as 170, the conversion looks correct. Now lets convert it to integer below   </span>

```python
df["Duration"] = df["Duration"].astype("int")
```

```python
df.info()
```

<span style='background:yellow'>**Observation:** Duration is now an Integer column. We can now sort based on the total minutes, high, low etc  </span>

## Finding Best Random State

```python
maxAccu=0
BestRs=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)    
    mod=LinearRegression()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        BestRs=i
print("Best accuracy is ", maxAccu,"on Random_state  ",BestRs)
```

<span style='background:yellow'>**Observation:** Now we found the best Random state, lets use it in our data  </span>

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=BestRs)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Instantiating the Algos

```python
lm=LinearRegression()
ridge = Ridge()
gb= GradientBoostingRegressor()
dt  = DecisionTreeRegressor()
rf  = RandomForestRegressor()
knn = KNeighborsRegressor()
ls=Lasso()
svr=SVR()
br = BaggingRegressor()
```

## Building the Model & Evaluation Metrics

```python
modelList=[("Lin Reg", lm), ("Lasso",ls),("Ridge", ridge),("Gradient boost",gb),("Bagging Reg",br),("Decision Tree", dt),("Random Forest",rf),("Kneigbor",knn)]

def GetAccuracy(x,y,modelList):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=BestRs)
    modelerror=[]
    ModelName= []
    modelR2=[]
    modelMAE=[]
    modelMSE=[]
    modelRMSE=[]
    modelRMSLE=[]
    modelCrossval=[]
    R2CVDiff=[]
    modelCrossval1=[]
    RMSECVDiff=[]
    modelCrossval2=[]
    RMSLECVDiff=[]
    
    for Nameofmodel, modelInstance in modelList:
        ModelName.append(Nameofmodel)
        modelInstance.fit(x_train,y_train)
        pred=modelInstance.predict(x_test)
        error=modelInstance.score(x_train,y_train)
        modelerror.append(error)
        r2=r2_score(y_test,pred)
        modelR2.append(r2)
        mae = mean_absolute_error(y_test,pred)
        modelMAE.append(mae)
        MSE=mean_squared_error(y_test,pred)
        modelMSE.append(MSE)
        RMSE=sqrt(mean_squared_error(y_test,pred))
        modelRMSE.append(RMSE)
        RMSLE=sqrt(mean_squared_log_error(y_test,pred))
        modelRMSLE.append(RMSLE)
        
        crossAcc=cross_val_score(modelInstance,x,y,cv=(KFold(n_splits=15,shuffle=True)),scoring='r2').mean()
        crossAcc1=cross_val_score(modelInstance,x,y,cv=(KFold(n_splits=15,shuffle=True)),scoring='neg_root_mean_squared_error').mean()
        crossAcc2=cross_val_score(modelInstance,x,y,cv=(KFold(n_splits=15,shuffle=True)),scoring='neg_mean_squared_log_error').mean()
        modelCrossval.append(crossAcc)
        R2CVDiff.append(r2-crossAcc)
        modelCrossval1.append(crossAcc1)
        RMSECVDiff.append(RMSE+crossAcc1)
        modelCrossval2.append(crossAcc2)
        RMSLECVDiff.append(RMSLE+crossAcc2)
    df_acc=pd.DataFrame({"Name":ModelName,
                         "Error":modelerror,
                         "R2": modelR2,
                         "MAE": modelMAE,
                         "MSE": modelMSE,
                         "RMSE": modelRMSE,
                         "RMSLE": modelRMSLE,
                         "R2-CV":modelCrossval,
                         "R2CVDiff":R2CVDiff,
                         "RMSE-CV":modelCrossval1,                         
                         "RMSECVDiff":RMSECVDiff,
                         "RMSLE-CV":modelCrossval2,
                         "RMSLECVDiff":RMSLECVDiff
                          })
    return df_acc
GetAccuracy(x,y,modelList)
    
    
```

<span style='background:yellow'>**Observation:** We can use teh R2 and RMSLE as our evaluation metrics, for the 8 algorithims we have chosen. RMSE punishes the error very high. We have also calculated the respective CV and found the diff between R2 &R2Cv, same for RMSE and RMSLE. When we go to R2, we need a high score, we can see that in Random forest and Gradient boost. We cannot go by best diff here, coz Regression models have low score. Lets use Gradient Boost and Random forest for our Hyper Parameter Tuning  </span>
