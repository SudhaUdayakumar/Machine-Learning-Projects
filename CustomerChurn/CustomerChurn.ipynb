# Problem Statement (Given)

Customer Churn Analysis
Problem Statement:
Customer churn is when a company’s customers stop doing business with that company. Businesses are very keen on measuring churn because keeping an existing customer is far less expensive than acquiring a new customer. New business involves working leads through a sales funnel, using marketing and sales budgets to gain additional customers. Existing customers will often have a higher volume of service consumption and can generate additional customer referrals.

Customer retention can be achieved with good customer service and products. But the most effective way for a company to prevent attrition of customers is to truly know them. The vast volumes of data collected about customers can be used to build churn prediction models. Knowing who is most likely to defect means that a company can prioritise focused marketing efforts on that subset of their customer base.

Preventing customer churn is critically important to the telecommunications sector, as the barriers to entry for switching services are so low. 

You will examine customer data from IBM Sample Data Sets with the aim of building and comparing several customer churn prediction models.

# Model Building (Classification)

# Importing Libraries for Model Building (Classification)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

## x & y data

```python
df_x=df.drop("Churn",axis=1)
y=df["Churn"]
```

## Min Max Scalar

```python
from sklearn.preprocessing import MinMaxScaler
```

<span style='background:pink'>**Note:** data range is not high, so we are going for MinMax scalar  </span>


```python
#diff between min max & Standdard sclar
#diff is in the range
#min-max- rescales data, all the feature values are between 0 and 1
#SS- it removes the mean, and scales the data to unit variance. 
```

```python
mms=MinMaxScaler()
mms.fit(df_x)
x=mms.transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

## Train & Test Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=43)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_test.head()
```

## Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
random_over_sampler = RandomOverSampler()
```

```python
x_res,y_res=random_over_sampler.fit_resample(x,y)
```

<span style='background:pink'>**Note:** Since it is an imbalanced data, we are using Over Sampling  </span>

## Best Random State

```python
maxAccu=0
BestRs=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)    
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    f1=f1_score(y_test,pred,average="macro")
    if f1>maxAccu:
        maxAccu=f1
        BestRs=i
print("Best accuracy is ", maxAccu,"on Random_state  ",BestRs)
```

<span style='background:yellow'>**Observation:** we have got the best random state, let us use it in our model   </span>

## Instantiating the Models

```python
lg=LogisticRegression()
gnb=GaussianNB()
sv=SVC()
rf=RandomForestClassifier(n_estimators=100)
ad=AdaBoostClassifier(n_estimators=100)
dtc=DecisionTreeClassifier()
```

## Model Building and checking Evaluation Scores

```python
model=[lg,gnb,rf,ad,dtc,sv]

for m in model:
    m.fit(x_res,y_res)
    pred=m.predict(x_res)
    print("\n f1 score: of ",m)
    print(f1_score(y_res,pred))
    print("\n  Classificatin report:  of ",m)
    print(classification_report(y_res,pred))
    print("***************************************")
    score=cross_val_score(m,x_res,y_res,scoring="f1",cv=5)
    print(score)
    print("mean of cross val with scoring= accuarcy:",score.mean())
    print("***************************************")
    
```

<span style='background:yellow'>**Observation:** We have 6 models, and we have done the F1 score, Classification report and the Cross validation. This was run especially to see the F1 score of 0 & 1 class. It is almost equal in all models. So this shows that are data is balanced. Decision tree & Random Forest tops our chart. Lets closely look at at F1 score and Cross validation in the next step   </span>

# Model Building and checking the Evaluation Scores

```python
modelList=[("Log Reg",lg),("Ada boost",ad),("Dec Tree Classifier", dtc),("Random Forest Class",rf),("Naive Bayes-G", gnb),("Support Vector",sv)]

def GetAccuracy(x,y,modelList):
    x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=.20,random_state=BestRs)
    ModelName= []
    modelf1=[]
    modelCrossval=[]
    f1CVDiff=[]
    for Nameofmodel, modelInstance in modelList:
        ModelName.append(Nameofmodel)
        modelInstance.fit(x_train,y_train)
        pred=modelInstance.predict(x_test)
        f1score= f1_score(y_test,pred,average="macro")
        modelf1.append(f1score)
        crossval=cross_val_score(modelInstance,x_res,y_res,scoring="f1",cv=5).mean()
        modelCrossval.append(crossval)
        f1CVDiff.append(f1score-crossval)
    df_acc=pd.DataFrame({"Name":ModelName,
                         "f1": modelf1,
                         "CrossValidation":modelCrossval,
                          "Difference in Accuracy": f1CVDiff})
    return df_acc
GetAccuracy(x_res,y_res,modelList)
```
